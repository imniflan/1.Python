```python
#Body Mass Index
bmi = float(input("Enter the BMI Index = "))
if(bmi > 35):
    print("Extremely Obese")
elif(bmi >= 30):
    print("Obese")
elif(bmi >= 25):
    print("Over Weight")
elif(bmi >= 18.5):
    print("Normal")
else:
    print("Under Weight")
```
